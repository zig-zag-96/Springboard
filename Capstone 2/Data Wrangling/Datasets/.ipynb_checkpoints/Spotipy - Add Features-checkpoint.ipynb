```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import numpy as np


#https://spotipy.readthedocs.io/en/2.19.0/
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id=API_CLIENT_ID,client_secret=API_CLIENT_SECRET)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

#sp.audio_features({'URI'}) gives us a dictionary within a list of the features and their values
#testing how this works

def GetAudioFeatures(URI):
    features = sp.audio_features(URI)
    #df_features = pd.DataFrame.from_dict(features)
    #dataframe = dataframe.merge(df_features, how='inner', left_on='URI', right_on='id')
    return features[0]
 
```

```python
def split_dataframe(df, chunk_size = 100): 
    chunks = list()
    num_chunks = len(df) // chunk_size + 1
    for i in range(num_chunks):
        chunks.append(df[i*chunk_size:(i+1)*chunk_size])
    return chunks
```

```python
for chunks in split_dataframe(billboard_songs):
    try:
        GetAudioFeatures(chunks['URI'], chunks)
    except:
        print('error')
```
